```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import pickle 
```

```python
def oneHotEncodeY(Y, nb_classes):
	return (np.eye(nb_classes)[Y]).astype(int)
```

```python
train = pickle.load(open("xor/train.pkl", 'rb'))
test = pickle.load(open("xor/test.pkl", 'rb'))
print(test)
testX, testY = np.array(test[0]), np.array(oneHotEncodeY(test[1],2))
trainX, trainY = np.array(train[0][:7000]), np.array(oneHotEncodeY(train[1][:7000],2))
valX, valY = np.array(train[0][7000:]), np.array(oneHotEncodeY(train[1][7000:],2))

trainX=torch.Tensor(trainX)
trainY=torch.Tensor(trainY)
testX=torch.Tensor(testX)
testY=torch.Tensor(testY)
```

```python
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
       
        self.fc1 = nn.Linear(2, 32)  # 6*6 from image dimension
        self.fc2 = nn.Linear(32, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
#         print("---------------------",x)
        return x



net = Net()
print(net)
optimizer = optim.SGD(net.parameters(), lr=0.00001)
```

```python
def prog():
    optimizer.zero_grad()
    for ip,tr in zip(trainX, trainY):
        out = net(ip)
        loss_function = nn.MSELoss()
        loss=loss_function(out, tr)
        loss.backward()
        optimizer.step()
```

```python
def accuracy():
#     count = 0
#     for a,b in zip(testX, testY):
#         print(net(a).data,b)
# #         if np.argmax(net(a)) == np.argmax(b):
# #             count=count+1
# #     print(count)
# #     print(count/2000)
    with torch.no_grad():
        out=net(testX)
        accur = (torch.softmax(out, dim=1).argmax(dim=1) == testY.argmax(dim=1)).sum().float() / float( testY.size(0) )
        return(accur)
```

```python
for i in range(20):
    prog()
    print("accu",i,accuracy()*100)
    
    
```
